● Pnadas是Python的協力廠商函式庫，提供資料框架等獨特資料以及各式各樣的資料處理功能，式解析資料時不可或缺的工具。

## <font color=red>匯入檔案 
    
-----
   

```python
import pandas as pd

df = pd.read_csv('nba.csv')
df
```

## <font color=red>檢視資料
    
-----
   

```python
#檢視是否有缺失值
df.isnull().sum()
#填入缺失值
#df = df.fillna(num)
```

```python
#清除缺失值
df = df.dropna() 
```

```python
print('---資料維度---')
print(df.ndim)
print('---欄列個數---')
print(df.shape)
print('---資料類型---')
print(df.dtypes)
print('---頭5筆---')
print(df.head())
print('---尾5筆---')
print(df.tail())
print('---名稱,筆數,大小,資料型態等---')
print(df.info)
```

## <font color=red>顯示資料--索引方式
    
-----
   

```python
df['Name']#用欄名稱，顯示這一欄的index名稱、內容及資料型態
```

## <font color=red>顯示排序
    
-----
   

```python
print(df.sort_values('Age'))
```

```python
df.sort_values('Name',ascending=False) #大到小
```

## <font color=red>   資料篩選
    
-----

&與 &~不與

篩選數量不限，想放幾個都可以

```python
df[(df['Age']>=25) & (df['Position']=='PF') &~ (df['Number']>=20)]
```

```python
df.loc[417,'Team'] #先選row 再選col
```

```python
df.loc[217:300,'Name'] #取217~300的'Name'資料
```

```python
df.loc[:,'Name':'Number'] #選取Name~Number所有資料
```
